```python
import matplotlib.pyplot as plt

y = [1, 2, 3, 4]
x = [9, 11, 13, 10]
plt.plot(x,y)

plt.ylabel("y")
plt.xlabel("x")
plt.title("Kevin Gates")
```

```python
fig, ax = plt.subplots()

colors = ['pink', 'blue', 'green', 'purple']
people = [56, 40, 10, 15]
bar_labels = ['pink', 'blue', 'green', 'purple']
bar_colors = ['tab:pink', 'tab:blue', 'tab:green', 'tab:purple']

ax.bar(colors, people, label=bar_labels, color=bar_colors)

ax.set_ylabel('fav color')
ax.set_title('most favored colors')
ax.legend(title='colors')
```

```python
labels = 'netflix', 'hulu', 'youtube', 'prime video'
sizes = [15, 50, 20, 15]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```
