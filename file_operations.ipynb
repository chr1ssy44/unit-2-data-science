*Use the code block below to open the same file, using a `for` loop print each line to the console.*

````python
statesFile = open('states.txt')
with open('states.txt') as statesFile:
    contents = statesFile.readlines()
# I made a variable 'statesFile' and opened the 'states.txt' file. Then I used 'with' to open the file again, I made another variable 'contents' to indicate I want to read thr file. Finally I printed the variable 'contents' to display the file.
for x in contents:
    print(x)
````

3. How would we need to transform the data we have, to be displayed in this chart?

```
We would need to use matplotlib to be able to present 'states.txt' as a bar graph.
```

Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt

one = 0
two = 0
three = 0
four = 0
five = 0


for x in contents:
    if x == "FL\n":
        one += 1
    elif x == "CA\n":
        two += 1
    elif x == "TX\n":
        three += 1
    elif x == "IN\n":
        four += 1
    elif x == "NY\n":
        five += 1

my_list = [one,two,three,four,five]

plt.bar(["FL", "CA", "TX", "IN", "NY"], my_list)
plt.xlabel("States")
plt.ylabel("Times Listed")
````
