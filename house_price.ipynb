What is the average pricing throughout the years?

```python
import pandas as pd
import matplotlib.pyplot as plt
print(plt.style.available)
plt.style.use("fivethirtyeight")

prices = pd.read_csv('housing_price_dataset.csv')
prices.head()

x = prices.groupby(['YearBuilt']).all()
print(x)
data1 = prices.groupby(['YearBuilt'])[['Price']].mean()
print(data1)

data1.plot(kind='line')
plt.ylabel("Average Prices")


```

This graph clearly shows the correlation between the years each house was built and the average pricing. The x-axis is the years ftom 1950 - 2020 when the houses when built and the y-axis in the average pricing ranging from 218,000 - 230,000.

What is the total amount of bathrooms made each year?

```python
plt.style.use("bmh")

a = prices.groupby(['Bathrooms']).all()
print(a)


data2 = prices.groupby(['YearBuilt'])[['Neighborhood','Bathrooms']].count()
print(data2)

data2.plot(kind='line')
plt.ylabel("Total Amount of Bathrooms")
```

This graph clearly shows the correlation between the years each house was built and the amount of bathrooms bulit in each house. The x-axis is the years ftom 1950 - 2020 when the houses when built and the y-axis in the amount of bathrooms bulit.
